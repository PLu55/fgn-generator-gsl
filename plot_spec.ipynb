```python
import numpy as np
import matplotlib.pyplot as plt

def load_data(file):
    with open(file, 'r') as f:
        l = []
        line = f.readline()
        while line:
            l.append(float(line))
            line = f.readline()
    return np.array(l)


pspec = load_data('pspec.txt')
out = load_data('out.txt')
#plt.plot(pspec)
plt.plot(out)
plt.show()
out = out - np.mean(out)
p = np.square(np.abs(np.fft.rfft(out)))
plt.plot(p)
plt.show()
```

```python
H = 0.75
n = pspec.shape[0]-1
N = 2*n
data = np.zeros((N,), dtype=np.complex64)

rnd=np.random.normal(size=N+1)
#rnd=rnd.reshape(rnd.shape[0],1)
data[0] = np.sqrt(2.0 * (np.power(N, 2.0*H) - np.power((N-1), 2.0*H)))*rnd[0]
data[n] = np.sqrt(2.0 * pspec[n])*rnd[N]

pspecsqrt = np.sqrt(pspec)
data[1:n] = pspecsqrt[1:n]*(rnd[1:N-1:2] + rnd[2:N-1:2]*1.j)
data[n+1:N] = np.conj(data[n-1:0:-1])
out=np.fft.ifft(data)
plt.plot(out.real[0:n])
plt.show()
plt.plot(np.cumsum(out.real[0:n]))
plt.show()
p = np.square(np.abs(np.fft.rfft(out.real[0:n])))
plt.plot(p)
plt.show()
```
